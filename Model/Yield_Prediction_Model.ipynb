```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
maize_yield = pd.read_csv("extracted_data/cybench-data/maize/IN/yield_maize_IN.csv")
wheat_yield = pd.read_csv("extracted_data/cybench-data/wheat/IN/yield_wheat_IN.csv")

maize_weather = pd.read_csv("extracted_data/cybench-data/maize/IN/meteo_maize_IN.csv")
wheat_weather = pd.read_csv("extracted_data/cybench-data/wheat/IN/meteo_wheat_IN.csv")

maize_soil = pd.read_csv("extracted_data/cybench-data/maize/IN/soil_moisture_maize_IN.csv")
wheat_soil = pd.read_csv("extracted_data/cybench-data/wheat/IN/soil_moisture_wheat_IN.csv")
```

```python
for df in [maize_yield, wheat_yield]:
    df['harvest_date'] = pd.to_datetime(df['harvest_year'].astype(str) + "-12-31")

# Convert weather and soil data dates
for df in [maize_weather, maize_soil, wheat_weather, wheat_soil]:
    df['date'] = pd.to_datetime(df['date'])
```

```python
def merge_closest(yield_df, weather_df, soil_df):
    weather_df = weather_df.sort_values('date')
    soil_df = soil_df.sort_values('date')
    yield_df = yield_df.sort_values('harvest_date')

    yield_weather = pd.merge_asof(yield_df, weather_df, left_on='harvest_date', right_on='date', by='adm_id', direction='backward')
    full_data = pd.merge_asof(yield_weather, soil_df, left_on='harvest_date', right_on='date', by='adm_id', direction='backward')

    return full_data
```

```python
maize = merge_closest(maize_yield, maize_weather, maize_soil)
wheat = merge_closest(wheat_yield, wheat_weather, wheat_soil)
```

```python
from xgboost import XGBRegressor # import the XGBRegressor
from sklearn.model_selection import RandomizedSearchCV # import RandomizedSearchCV


# Define features and target
features = ['tmin', 'tmax', 'prec', 'rad', 'tavg', 'et0', 'cwb', 'ssm', 'rsm', 'crop_name', 'year', 'month', 'previous_yield', 'prec_tavg', 'et0_cwb_ratio']
target = 'yield'

# Drop rows with missing target values
data = data.dropna(subset=[target])

# Split data
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target], test_size=0.2, random_state=42)

# Scale features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Hyperparameter tuning with RandomizedSearchCV
param_dist = {
    'n_estimators': [100, 200, 300, 400],
    'max_depth': [3, 5, 7, 10],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0]
}

model = XGBRegressor(random_state=42)
random_search = RandomizedSearchCV(model, param_dist, n_iter=30, cv=3, n_jobs=-1, verbose=2, random_state=42)
random_search.fit(X_train, y_train)
```

```python
# Best model
best_model = random_search.best_estimator_

# Predictions
y_pred = best_model.predict(X_test)

# Evaluate
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print(f'Optimized RMSE: {rmse}, Optimized R2 Score: {r2}')
```

```python
from sklearn.ensemble import StackingRegressor
from lightgbm import LGBMRegressor
```

```python



#  Define features and target
features = ['tmin', 'tmax', 'prec', 'rad', 'tavg', 'et0', 'cwb', 'ssm', 'rsm', 'crop_name', 'year', 'month', 'previous_yield', 'prec_tavg', 'et0_cwb_ratio']
target = 'yield'

# Drop rows with missing target values
data = data.dropna(subset=[target])

# Split data
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target], test_size=0.2, random_state=42)

# Scale features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Base models
xgb = XGBRegressor(n_estimators=300, max_depth=7, learning_rate=0.05, subsample=0.8, colsample_bytree=0.8, random_state=42)
lgbm = LGBMRegressor(n_estimators=300, learning_rate=0.05, max_depth=7, random_state=42)

# Stacking model
stacking_model = StackingRegressor(
    estimators=[('xgb', xgb), ('lgbm', lgbm)],
    final_estimator=XGBRegressor(n_estimators=100, learning_rate=0.05, max_depth=5, random_state=42),
    passthrough=True
)

# Train the model
stacking_model.fit(X_train, y_train)
```

```python
# Train the model
stacking_model.fit(X_train, y_train)

# Predictions
y_pred = stacking_model.predict(X_test)

# Evaluate
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print(f'Optimized RMSE: {rmse}, Optimized R2 Score: {r2}')
```

```python

from sklearn.impute import KNNImputer
# Feature Engineering: Adding time-based features
data['year'] = pd.to_datetime(data['harvest_date']).dt.year
data['month'] = pd.to_datetime(data['harvest_date']).dt.month
data['previous_yield'] = data.groupby('adm_id')['yield'].shift(1)

# Additional interaction features
data['prec_tavg'] = data['prec'] * data['tavg']
data['et0_cwb_ratio'] = data['et0'] / (data['cwb'] + 1e-6)

# Handle missing values using KNN Imputer
# Selecting numeric columns for imputation
numeric_cols = data.select_dtypes(include=[np.number]).columns
data[numeric_cols] = KNNImputer(n_neighbors=5).fit_transform(data[numeric_cols])

# Define features and target
features = ['tmin', 'tmax', 'prec', 'rad', 'tavg', 'et0', 'cwb', 'ssm', 'rsm', 'crop_name', 'year', 'month', 'previous_yield', 'prec_tavg', 'et0_cwb_ratio']
target = 'yield'

# Drop rows with missing target values
data = data.dropna(subset=[target])

# Split data
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target], test_size=0.2, random_state=42)

# Scale features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Base models
xgb = XGBRegressor(n_estimators=500, max_depth=10, learning_rate=0.03, subsample=0.8, colsample_bytree=0.8, random_state=42)
lgbm = LGBMRegressor(n_estimators=500, learning_rate=0.03, max_depth=10, num_leaves=31, random_state=42)

# Stacking model
stacking_model = StackingRegressor(
    estimators=[('xgb', xgb), ('lgbm', lgbm)],
    final_estimator=XGBRegressor(n_estimators=150, learning_rate=0.03, max_depth=6, random_state=42),
    passthrough=True
)

# Train the model
stacking_model.fit(X_train, y_train)

# Predictions
y_pred = stacking_model.predict(X_test)

# Evaluate
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print(f'Optimized RMSE: {rmse}, Optimized R2 Score: {r2}')
```
